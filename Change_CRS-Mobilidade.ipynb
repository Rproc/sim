```python
import pandas as pd
from pyproj import Proj, transform
import fiona
from fiona.crs import from_epsg
import logging
import sys


# outProj = Proj(init='epsg:3857')
# inProj = Proj(init='epsg:4326') # It is equal to WSG
```

```python
shape = fiona.open("Mapas_sim/env/trans_onibus_por_embarque.shp")
print (shape.schema)
```

```python
for i in range(0, len(shape)):
    print((shape[i]['geometry']['type']))
```

```python
# Transformar escala lag, long em Metros no globo achatado

# salvando o esquema e pegando a projecao de entrada
sink_schema = shape.schema.copy()
p_in = Proj(shape.crs)

# definindo um novo shape de saida, no mesmo local do shape atual com a projeção mudada
with fiona.open(
        'ponto_onibus_metros.shp', 'w',
        crs=from_epsg(3857),
        driver=shape.driver,
        schema=sink_schema,
        ) as sink:

    p_out = Proj(sink.crs)

    for f in shape:
        try:
            # verifica se eh da estrutura poligono
            assert f['geometry']['type'] == "MultiPoint"
            l = []
            # para todas as coordenadas do poligono
            for j in f['geometry']['coordinates']:
                l.append(j)

            new_coords = []
            for elem in l:
                x2, y2 = transform(p_in, p_out, elem[0], elem[1]) #transforma
                new_coords.append((x2, y2))
            f['geometry']['coordinates'] = new_coords
            l = []
            sink.write(f)

        except Exception:
        # Writing uncleanable features to a different shapefile
        # is another option.
            logging.exception("Error transforming feature %s:", f['id'])
```
